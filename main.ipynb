```python

from ebaysdk.finding import Connection as Finding
from bs4 import BeautifulSoup
from dotenv import load_dotenv
import os
import pandas as pd

load_dotenv()

# eBay API endpoint
EBAY_API_URL = "https://api.ebay.com/buy/browse/v1/item_summary/search"

# Get OAuth token from environment variable
OAUTH_TOKEN = os.getenv("EBAY_OAUTH_TOKEN")

# Function to fetch eBay data for Apple Watches
def fetch_ebay_data(query: str, limit: int = 100):
    api = Finding(appid="MosesMac-pricemon-PRD-85bb4ae7d-e6d9b5f3", config_file=None)
    response = api.execute('findItemsAdvanced', {'keywords': query, 'paginationInput': {'entriesPerPage': limit}})
    soup = BeautifulSoup(response.content, 'lxml')
    items = soup.find_all('item')
    relevant_data = []
    for item in items:
        item_info = {
            "Title": item.title.string,
            "Price": item.currentprice.string,
            "Country of Origin": item.location.string,
            "Product Link": item.viewitemurl.string
        }
        relevant_data.append(item_info)
    return relevant_data 

# Function to arrange price in ascending order
def arrange_price_ascending(data):
    data = sorted(data, key=lambda x: float(x["Price"].replace("$", "").replace(",", "")))
    return data

# Print the data
data = fetch_ebay_data("apple watch", 100)
sorted_data = arrange_price_ascending(data)
for i, item in enumerate(sorted_data[:10], start=1):
    print(f"{i}. Title: {item['Title']}")
    print(f"   Price: {item['Price']}")
    print(f"   Country of Origin: {item['Country of Origin']}")
    print(f"   Product Link: {item['Product Link']}")
    print()

# Save the data to Excel file
def save_to_excel(data, filename):
    df = pd.DataFrame(data)
    df.to_excel(filename, index=False)
    print(f"Apple Watch data saved to {filename}")

# Save the data to Excel file
save_to_excel(sorted_data, "apple_watch_data14.xlsx")
```
